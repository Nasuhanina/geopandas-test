```python
import json

# for data in georoute['features']:   
#     # print(data)
#     parsed_data = json.loads(data)
#     print(parsed_data)

# Create an empty list to store DataFrame
dfs = []

# Split the JSON object by index and convert each dictionary into a DataFrame
for record in georoute['features']:
    df = pd.DataFrame([record])  # Create a DataFrame from each record
    dfs.append(df)  # Append the DataFrame to the list

# Concatenate the list of DataFrames into a single DataFrame
result_df = pd.concat(dfs, ignore_index=True)

# Now you have a DataFrame with each record from the JSON object as a separate row
print(result_df)
```
