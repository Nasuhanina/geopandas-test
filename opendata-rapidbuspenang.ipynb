```python
# pip install gtfs-realtime-bindings pandas requests
from google.transit import gtfs_realtime_pb2
from google.protobuf.json_format import MessageToDict
import pandas as pd
from requests import get
 
# Sample GTFS-R URL from Malaysia's Open API
URL = 'https://api.data.gov.my/gtfs-realtime/vehicle-position/prasarana?category=rapid-bus-penang'
# Parse the GTFS Realtime feed
feed = gtfs_realtime_pb2.FeedMessage()
response = get(URL)
feed.ParseFromString(response.content)
 
# Extract and print vehicle position information
vehicle_positions = [MessageToDict(entity.vehicle) for entity in feed.entity]
print(f'Total vehicles: {len(vehicle_positions)}')
df = pd.json_normalize(vehicle_positions)
print(df.head())
```

```python
df
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'],unit='s')
```

```python
trip = pd.read_csv('rapidbuspenang-data/trips.txt')
shape = pd.read_csv('rapidbuspenang-data/shapes.txt')
stoptime = pd.read_csv('rapidbuspenang-data/stop_times.txt')
route = pd.read_csv('rapidbuspenang-data/routes.txt')
stop = pd.read_csv('rapidbuspenang-data/stops.txt')
```

```python
# trip
# shape
```

```python
newdata = pd.concat([df,trip],axis=1,join='inner')
newdata
```

```python
newdata = pd.concat([newdata,shape],axis=1, join='inner')
newdata
```

```python
newdata = pd.concat([newdata,stoptime],axis=1, join='inner')
newdata
```

```python
newdata = pd.concat([newdata,route],axis=1, join='outer')
newdata
```

```python
newdata = pd.concat([newdata,stop],axis=1, join='inner')
newdata
```

```python
newdata['timestamp'] = newdata['timestamp'].astype(str)
```

```python
from keplergl import KeplerGl
from datetime import datetime
import json

# Assuming the 'arrival_time' column represents the time dimension
# You can also use 'departure_time' or any other time-related column from your GTFS data

# Create a Kepler.gl map
map_1 = KeplerGl()

# Add data to the map
# map_1.add_data(data=station[station['coordinates']], name='coordinate')
map_1.add_data(data=newdata, name='rapid rail kl')

# Configure the Kepler.gl map to animate based on the timestamp column
map_1.config = {
    'version': 'v1',
    'config': {
         'mapState': {
            'latitude': 5.425300,
            'longitude': 100.312386,
            'zoom' : 11
        }
    }
}

# Visualize the animated data within Kepler.gl

# map_1
# You can further configure the map interactively in the Jupyter Notebook environment

# If you want to save the map configuration for later use or share it with others
# you can save it as a JSON file
# config = map_1.config

# configsavepenang = json.dumps(config, default=str)
# print(configsavepenang)
# with open('keplergl_config.json', 'w') as f:
#     f.write(configsavepenang)

# # You can also export the map as an HTML file
# map_1.save_to_html(file_name='rapidbuspenang.html')
map_1
```
