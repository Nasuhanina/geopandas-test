```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("19.06.16_travels_Bonn.csv")
data.head()
```

```python
data.rename(columns={'TAA': 'Alerts'}, inplace=True)
data.rename(columns={'TA': 'Delay hour'}, inplace=True)
data.rename(columns={'TIN': 'Train Model'}, inplace=True)
data.rename(columns={'TSI': 'StationID'}, inplace=True)
data.rename(columns={'TIR': 'Route'}, inplace=True)
data.rename(columns={'TIM': 'Direction Dept/Ar'}, inplace=True)
data.rename(columns={'TIL': 'Req w parameters'}, inplace=True)
data.rename(columns={'TIRE': 'Dest'}, inplace=True)
data.rename(columns={'TIP': 'Platform number'}, inplace=True)
data.rename(columns={'TIT': 'Dept hour'}, inplace=True)
data.rename(columns={'TID': 'Date'}, inplace=True)
data.rename(columns={'TSC': 'Station Name and ID'}, inplace=True)
data.rename(columns={'TAc': 'Delay in min'}, inplace=True)
```
